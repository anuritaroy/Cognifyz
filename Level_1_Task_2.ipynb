```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\amiro\\Desktop\\Internship cognifiz\\Dataset .csv")
```

```python
# Display the first few rows to verify loading
df.head()
```

```python
# Select numerical columns
numerical_cols = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']
```

```python
# Calculate basic statistics
stats = df[numerical_cols].describe()
```

```python
print(stats)
```

```python
# Add median explicitly (describe() includes mean, std, etc., but median is the 50% mark)
stats.loc['median'] = df[numerical_cols].median()
```

```python
print(stats)
```

```python
# Display the statistics
print("Basic Statistical Measures for Numerical Columns:")
print(stats)
```

```python
#Visualize distributions with histograms
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Country Code distribution
country_dist = df['Country Code'].value_counts(normalize=True) * 100  # Percentage
print("\nCountry Code Distribution (%):")
print(country_dist)
```

```python
# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x=country_dist.index, y=country_dist.values)
plt.title('Distribution of Country Code')
plt.xlabel('Country Code')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# City distribution (top 10 for clarity)
city_dist = df['City'].value_counts().head(10)  # Top 10 cities
print("\nTop 10 Cities by Restaurant Count:")
print(city_dist)
```

```python
# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x=city_dist.index, y=city_dist.values)
plt.title('Top 10 Cities by Restaurant Count')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
# Split cuisines and count individual occurrences
cuisines_series = df['Cuisines'].dropna().str.split(', ').explode()
cuisine_dist = cuisines_series.value_counts().head(10)  # Top 10 cuisines
print("\nTop 10 Cuisines by Frequency:")
print(cuisine_dist)
```

```python
# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x=cuisine_dist.index, y=cuisine_dist.values)
plt.title('Top 10 Cuisines by Frequency')
plt.xlabel('Cuisine')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
